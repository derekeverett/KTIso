```python
def calc_v2(e,ux,uy):
    """Calculates a pseudo observable for eliptic flow using the energy,and flowvelocity projections"""
    if e.shape == ux.shape == uy.shape:
        mom_x = np.multiply(e,ux)
        mom_y = np.multiply(e,uy)
        #mom_x = ux
        #mom_y = uy
        squre_diff = np.square(mom_x) - np.square(mom_y)
        squre_add = np.square(mom_x) + np.square(mom_y)
        v2 = np.divide(np.sum(squre_diff),np.sum(squre_add))
    else:
        raise valueError('Input matrices have to have the same dimensions')
    return v2
```
